```python
import csv
import pandas as pd
import numpy as np
```

```python
SchoolData = pd.read_csv('Rschools.csv')
StudentData = pd.read_csv('Rstudents.csv')
SchoolData.head()
```

```python
StudentData.head()
```

```python
SDcomplete = pd.merge(SchoolData, StudentData, how = "left", on = ["school_name"])
#SDcomplete = pd.merge(SchoolData, StudentData, how = "left", on = ["school_name", "school_name"])
SDcomplete.head()
```

```python
Schools = SDcomplete['school_name'].value_counts()
TNumbSchools = Schools.count()
TNumbStudents = SDcomplete['student_name'].count()

print(Schools)
print("------------------------------------------")
print("Number of students: " + str(TNumbStudents))
print("Number of schools: " + str(TNumbSchools))
```

```python
#Bufget per school
schoolsB = SDcomplete['budget'].value_counts()
schoolsB
ToBudget = SchoolData["budget"].sum()
print("Total Budget: " + str(ToBudget))
```

```python
MathS = SDcomplete['math_score']
AverageM = MathS.mean()
ReadS = SDcomplete['reading_score']
AverageR = ReadS.mean()
OverallS = (AverageM + AverageR)/2
print("Average math score: " + str(AverageM))
print("Average reading score: " + str(AverageR))
print("Overall passing rate: " + str(OverallS))
```

```python
PassingSM = SDcomplete.loc[SDcomplete["math_score"] >= 70, :]
PassingSM.head()
NumbPassSM = PassingSM['math_score'].count()
pm = (NumbPassSM*100)/TNumbStudents
PassingSR = SDcomplete.loc[SDcomplete["reading_score"] >= 70, :]
PassingSR.head()
NumbPassSR = PassingSR['reading_score'].count()
pr = (NumbPassSR*100)/TNumbStudents

print("Percent Passing Math: " + str(pm))
print("Percent Passing Reading: " + str(pr))
```

```python
result = pd.DataFrame({"Total schools":[TNumbSchools],
                        "Total Students":[TNumbStudents],
                       "Total Budget":[ToBudget],
                       "Average Math Score":[AverageM],
                       "Average Reading Score":[AverageR],
                       "% Passing Math":[pm],
                       "% Passing Reading":[pr],
                       "Overall Passing Rate":[OverallS]
                       })
result
```
